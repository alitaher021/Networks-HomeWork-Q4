```python
# Import Required modules
import networkx as nx
import matplotlib.pyplot as plt
import random
```

```python
# N number of nodes from user
#print("Enter number of nodes")
N = 10
  
# probability value for edges
#print("Enter value of probability of every node")
P = 0.35

def distribution_graph(g):
    print(nx.degree(g))
    all_node_degree = list(dict((nx.degree(g))).values())
  
    unique_degree = list(set(all_node_degree))
    unique_degree.sort()
    nodes_with_degree = []
    for i in unique_degree:
        nodes_with_degree.append(all_node_degree.count(i))
  
    plt.plot(unique_degree, nodes_with_degree)
    plt.xlabel("Degrees")
    plt.ylabel("No. of nodes")
    plt.title("Degree distribution")
    plt.show()
```

```python
# The graph code below has been taken from geeksforgeeks.
# Create an empty graph object
g = nx.Graph()

# Adding nodes
g.add_nodes_from(range(1, N + 1))  

average_degree = 0
average_clustering = 0
average_path_length = 0

for i in range(5):
    # Add edges to the graph randomly.
    for i in g.nodes():
        for j in g.nodes():
            if (i < j):
              
                # Take random number R.
                R = random.random()
              
                # Check if R<P add the edge to the graph else ignore.
                if (R < P):
                    g.add_edge(i, j)
        pos = nx.circular_layout(g)
      
        # Display the social network 
        nx.draw(g, pos, with_labels=1)
        plt.show()
    
    average_degree = average_degree + 2*g.number_of_edges() / float(g.number_of_nodes()) # Taken from stackoverflow.
    average_clustering = average_clustering + nx.average_clustering(g) # taken from geeksforgeeks.
    average_path_length = average_path_length + nx.average_shortest_path_length(g) # taken from networksx.org
    
    print('The average degree of this graph is:', 2*g.number_of_edges() / float(g.number_of_nodes()))
    print('The average clustering coeffiecient of this graph is: ',nx.average_clustering(g))
    print('The average path length of this graph is: ',nx.average_shortest_path_length(g))
    distribution_graph(g)

#distribution_graph(g)
print("The average degree after running 5 times: ", average_degree / 5)
print("The average clustering after running 5 times: ", average_clustering / 5)
print("The average path length after running 5 times: ", average_path_length / 5)

```
